```
!sudo apt install p7zip-full p7zip-rar

import re
import requests
import xml.etree.ElementTree as ET
import subprocess
import os.path

directory = "/content/podcast"
URL = 'https://feeds.megaphone.fm/stuffyoushouldknow'
resp = requests.get(URL)
body = resp.content.decode(encoding='utf-8')

subprocess.run(["mkdir", directory])

rss = ET.fromstring(body)
channel = rss[0]
for item in channel:
    if item.tag == 'item':
        title = re.sub(r'[\:\?\<\>\:\"\/\\\|\?\*]', '', item[0].text)
        filepath = directory +"/"+ title + ".mp3"
        if os.path.isfile(filepath):
            print (filepath + " already exist.")
        else:
            url = item[-1].attrib['url']
            try:
                subprocess.run(["wget", "-O", filepath, url])
            except (KeyboardInterrupt, SystemExit):
                subprocess.run(["rm", filepath])

!7z a SYSK.7z /content/podcast
```
